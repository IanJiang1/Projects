# Approximating Community Connection Probabilities in a Directed Graph using Gibbs Sampler

In the below, we assume a Bernoulli probability connection amongst $K$ different implied communities. Specifically, for each node $i$ we assume the following community index $z_i$ where $z_i \in [1, ..., K]$ lies in a finite discrete set. These are the variables of interest, whereas we are only given the a binary 0-1 value $x_{ij}$ for each edge which indicates whether or not an edge exists connecting the two. As the title implies, the algorithm  may be used to find communities within a directed graph. 

The following form is assumed for conditional probability of a node $i$ belonging to community $z_i$:


$
p(z_i | \boldsymbol{z}_{-i}, y, \boldsymbol{\pi}, W) = \prod_{j \neq i} W_{z_iz_j} ^ {\mathbb{I}(y_{ij} = 1)} (1 - W_{z_iz_j}) ^ {\mathbb{I} (y_{ij}= 0)}W_{z_jz_i} ^ {\mathbb{I} (y_{ji} = 1)} (1 - W_{z_jz_i}) ^ {\mathbb{I} (y_{ji} = 0)} \sum_{k = 1} ^ K \pi_k \mathbb{I} (z_i = k)
$

Therefore, the presece of an edge 'turns on' the associated probability inside this expression for the full conditional. The full conditionals for each of the parameters in the model are maximized in turn, iterating over and over until the total log likelihood is maximized. 

There are several different initial parameter values which we may choose in order to start the algorithm. In practice, we will randomly hold out several of the edges in the graph as a test set, using the remaining subgraph to train the algorithm. We proceed in this manner for several different initializations, and choose the one which maximizes the likelihood of the selected test subgraph. This performed in the 'testLawyer.py' code.

Obviously, the ability to find natural groupings within such a general graph framework has far-reaching implications in most fields of study, including biology and the social sciences>
