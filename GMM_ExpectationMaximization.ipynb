# Cluster Identification using EM Algorithm on Gaussian Mixture

Here we attempt to identify natural clusters in 2-dimensional data using the Expectation-Maximization algorithm, assuming that data is distributed as Gaussian Mixture. Specifically, data are assumed to belong to one of $K$ different clusters, and the likelihood of the data is given by

$
p(\boldsymbol{z}, X, \boldsymbol{\mu}, \Sigma) = \prod_{i = 1} ^ N \prod_{k = 1} ^ K \left[ \pi_k \mathcal{N} (x_i | \boldsymbol{\mu}_k, \Sigma_k) \right] ^ {\mathbb{I} (z_i = k)}
$

The expectation maximization algorithm proceeds as follows: we maximize the expectation of the likelihood, w.r.t to the full likelihood using current parameter estimates:

$
\boldsymbol{\theta}_{curr} = \text{argmax}_{\theta} \mathbb{E} \left[ p(X | \boldsymbol{\theta}) | \boldsymbol{\theta}_{old} \right]
$

By iteratively solving this equation, we eventually converge upon parameter estimates ($z_k, \mu_k, \Sigma_k $) which maximize the expected value of the likelihood.
